```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv')
df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Median'],y=df['Unemployment_rate'])
plt.xlabel('Median Salary')
plt.ylabel('Unemployment rate')
plt.title('correlation between Median Salary and Unemployment_rate')
#correlation 
correlation=df['Median'].corr(df['Unemployment_rate'])
print(f'correlation between Median Salary and Unemployment Rates:{correlation:.2f}')

plt.show()
```

```python
high_salary_low_Unemployment= df[(df['Median'] > 60000)&(df['Unemployment_rate'] < 0.05)]
print('Major with the highest salaries and low unemployment rates:')
high_salary_low_Unemployment
```

```python
#average grouped by major_catergory
grouped_data = df.groupby('Major_category')['Median'].mean()
print('Average Median Salary for each Major Category:')
grouped_data      
```

```python
#bar chart
grouped_data= df.groupby('Major_category')['Median'].mean()
plt.xlabel('Major category')
plt.ylabel('Average Median Salary')
plt.title('Average Median Salary by Major Category')
grouped_data.plot(kind='bar')

plt.show()
```
