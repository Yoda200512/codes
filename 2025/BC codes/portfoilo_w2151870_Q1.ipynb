```python
import matplotlib.pyplot as plt
#scatter plot
df = pd.DataFrame(data);   
# Draw a scatter plot 
df.plot.scatter(x = 'Price', y = 'Mileage', s = 1)
plt.show()
```

The scatter plot Mileage and price shows us that the car with a higher price tends to have a lower Mileage. This tells us that high priced cars are less used, while cars with lower prices tend to have a higher mileage. The relationship between mileage of cars and Price is an inverse relationship.

```python
mileage_price_corr = df['Mileage'].corr(df['Price'])
Engine_Size_price_corr = df['Engine_Size'].corr(df['Price'])
```
