```python
data = pd.read_csv('C:/Users/acer/Downloads/amazonFood.csv')
DF = pd.DataFrame(data)
DF
```

```python
DF['word_count'] = DF['Text'].str.len()
DF
```

```python
DF['length_category'] = ' '
#for loop
for index in DF.index:
    if DF.loc[index, 'word_count'] < 100:
        DF.loc[index, 'length_category'] = 'short'
    elif DF.loc[index, 'word_count'] > 200:
        DF.loc[index, 'length_category'] = 'long'
    else:
        DF.loc[index, 'length_category'] = 'medium'
DF
```

```python
#while loop
DF['length_category'] = ' '
# Initialize index
index = 0
#while loop
while index < len(DF.index):
    if DF.loc[index, 'word_count'] < 100:
        DF.loc[index, 'length_category'] = 'short'
    elif DF.loc[index, 'word_count'] > 200:
        DF.loc[index, 'length_category'] = 'long'
    else:
        DF.loc[index, 'length_category'] = 'medium'
    index +=1
DF
```

```python
#Vectorisation

DF['length_category'] = 'medium '
DF.loc[DF.word_count < 100,'length_category'] = 'short'
DF.loc[DF.word_count >200, 'length_category'] = 'long'
DF
```
